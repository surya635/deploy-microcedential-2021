```python
dummies = pd.get_dummies(df.lokasi)
lokasi_cat = pd.get_dummies(df.lokasi)
dummies.head()
```

```python
df = pd.concat([data,dummies],axis='columns')
df.head()
```

```python
df= df.drop(["Unnamed: 0"],axis=1)
df.head()
```

```python
df = df.drop(['lokasi'] , axis=1)
```

```python
y = df['harga']
X = df.drop(['harga','longitude'], axis = 1)

X
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
model = LinearRegression(normalize=True)
model.fit(X_train,y_train)

pred= model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
#Root Mean Square Error
from sklearn import metrics
import math
print(math.sqrt(metrics.mean_squared_error(y_test, pred)))
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import numpy as np
```

```python
algos = [LinearRegression(normalize=True),  Ridge(normalize=True), Lasso(normalize=True),
          KNeighborsRegressor(), DecisionTreeRegressor(),RandomForestRegressor(),XGBRegressor()]

names = ['Linear Regression', 'Ridge Regression', 'Lasso Regression',
         'K Neighbors Regressor', 'Decision Tree Regressor','Random Forest Regressor','XGBoost Regressor']

rmse_list = []
```

```python
for name in algos:
    model = name
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    MSE= metrics.mean_squared_error(y_test,pred)
    rmse = np.sqrt(MSE)
    rmse_list.append(rmse)
```

```python
evaluation = pd.DataFrame({'Model': names,
                           'RMSE': rmse_list})

evaluation
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error
```

```python
RF= RandomForestRegressor()
kf = KFold(n_splits=5, random_state=42, shuffle=True)
R2_scores = cross_val_score(RF, X_train, y_train, cv=kf,scoring = 'r2')
pred = cross_val_predict(RF, X_train, y_train, cv=kf)
mae= mean_absolute_error(y_train, pred)
```

Hyperparameter Tuning
We will try to improve the accuracy by tuning the hypermeters for this model. For that, we'll use Grid Search to get the optimised values of hypermeters.

We'll give the algorithm necessary options for the parameters to choose from and the algorithm will tune us the optimised set of parameter choices for the model.

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2,4]
# Minimum number of samples required to split a node
min_samples_split = [2, 5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV

kf = KFold(n_splits=5, random_state=42, shuffle=True)
model = RandomForestRegressor()

clf = GridSearchCV(estimator = model, param_grid = param_grid, cv = kf, verbose=True, n_jobs = 4)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
best_clf = clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
print (f'Train Accuracy - : {clf.score(X_train,y_train):.3f}')
print (f'Test Accuracy - : {clf.score(X_test,y_test):.3f}')
```
